```python
files = [f for f in os.listdir('players') if not f.startswith('.')]
agg_list = list()
for p in files:
    path = f"players/{p}"
    #print(path)
    temp_df = pd.read_csv(path)
    temp_df.drop(["Unnamed: 0"], axis=1, inplace = True)
    year_slice = temp_df.drop(temp_df[(temp_df.TEAM_ID != 0)].index)
    year_list = year_slice['SEASON_ID'].to_list() 
    temp_df = temp_df.drop(temp_df[(temp_df.SEASON_ID.isin(year_list)) & (temp_df.TEAM_ID != 0)].index)
    temp_df.reset_index()
    temp_df.to_csv(path)
    with open(f"players/{p}", newline='') as f:
        reader = csv.reader(f)
        data = list(reader)
        del data[0]
        for season in data:
            agg_list.append(season)
        
```

```python
test_player = pd.DataFrame(agg_list, columns=['', 'PLAYER_ID', 'SEASON_ID', 'LEAGUE_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'PLAYER_AGE', 
                                          'GP', 'GS', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 
                                          'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS'])
test_player
```

```python
year = 2023
url_list = list()
while year >= 1951:
    url = 'https://www.basketball-reference.com/allstar/NBA_{}.html'.format(year)
    url_list.append(url)
    year-=1


    
```

```python
from bs4 import BeautifulSoup  # library for HTML parsing
import time
year_list = list(range(1951, 2024))
year_list.reverse()
n = 0
allstar_dict = dict()
for url in url_list:
    temp_set = set()
    response = requests.get(url)
    html = BeautifulSoup(response.text, "html.parser")
    regex = '/players/'
    p = 0
    for a in html.find_all('a', href=True): 
        if re.search(regex, a['href']) and p < 2:
            if a.text == "Players":
                p+=1
            else:
                temp_set.add(a.text)
    l = list(temp_set)
    allstar_dict.update({year_list[n]: l}) 
    n+=1
    time.sleep(6)
```

```python
allstar_cat = ['year', 'player']
allstar_convert = list()
for year in year_list:
    for player in allstar_dict[year]:
        temp_list = [year, player]
        allstar_convert.append(temp_list)
allstar_df = pd.DataFrame(allstar_convert, columns=allstar_cat)
allstar_df.sort_values(by=['player', 'year'], inplace=True)
```

```python
allstar_df.to_csv('allstar_players.csv')
```
